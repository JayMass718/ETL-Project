```python
province_dfs=[]

for province in province_list:
    province_dfs.append(Sort_province[Sort_province['State']== province].sort_values(by=['Last Update'],axis=0,ascending=False).copy())

    
```

### Load to SQL

```python
from sqlalchemy import create_engine
```

```python
connection_string = "postgres:postgres@localhost:5432/nCoV_db"
```

```python
engine=create_engine(f"postgresql://{connection_string}")
```

```python
province_list[-1] = "Other"
fixed_cols = [column.lower().replace(" ","_") for column in province_dfs[0].columns]
```

```python
for virus_df,province_name in zip(province_dfs,province_list):
    virus_df.columns = fixed_cols
    virus_df.to_sql(name=province_name.lower().replace(" ","_"),con=engine,if_exists="replace",index=False)
```
